```python
def clean_location(df):
  df.dropna(subset=["Location"], inplace=True)

  df["Longitude"] = df["Location"].str.split(" ").apply(lambda s: s[1][1:])
  df["Latitude"] = df["Location"].str.split(" ").apply(lambda s: s[2][:-1])
```

```python
def clean_time(df):
  df.dropna(subset=["File Date"], inplace=True)

  df["File Date"] = pd.to_datetime(df["File Date"])
  df["Year"] = df["File Date"].apply(lambda x: x.year)
  df["Month"] = df["File Date"].apply(lambda x: x.month)
  df["Day"] = df["File Date"].apply(lambda x: x.day)
```

```python
clean_location(evictions)
```

```python
clean_time(evictions)
```

```python
reasons = []
for col in evictions.columns:
  if type(evictions[col][1]) == np.bool_:
    reasons.append(col)

evictions["Reason"] = ""
for row in evictions.index:
  for col in reasons:
    if evictions.loc[row, col] == True:
      evictions.loc[row, "Reason"] = col
      break;
```
